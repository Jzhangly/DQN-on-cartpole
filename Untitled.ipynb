```python

import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from collections import deque
import random
import gym
import matplotlib
import matplotlib.pyplot as plt
import itertools

# if gpu is to be used
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
class DQN(nn.Module):

    def __init__(self, n_observations, n_actions,epsilonS = 0.1, epsilonE = 0.1, epDecay = 0.98):
        super(DQN, self).__init__()
        self.nA = n_actions
        self.epsilon = epsilonS
        self.epsilonE = epsilonE
        self.epDecay = epDecay
        self.count = 1
        self.layer1 = nn.Linear(n_observations, 128)
        self.layer2 = nn.Linear(128, 128)
        self.layer3 = nn.Linear(128, n_actions)
        
    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        return self.layer3(x)
    
    def action(self,state):  
        if self.count % 100 == 0:
            if self.epsilon > self.epsilonE:
                self.epsilon = self.epsilon * self.epDecay
        self.count = self.count + 1
        
        A = np.ones(self.nA, dtype=float) * (self.epsilon/self.nA)
        temp = self.forward(state)
        best_action = temp.max(1)[1].data[0].item()
        A[best_action] += 1.0 - self.epsilon
        return A
```

```python
class replay_buffer(object):
    def __init__(self, capacity,nStep):
        self.nStep = nStep
        self.capacity = capacity
        self.memory = deque(maxlen=self.capacity)
        
    def store(self, state,action,reward,done,b_prob,sigma):
        state = state
        action = action
        reward = reward
        done = done
        b_prob = b_prob
        sigma = sigma
        self.memory.append([state,action,reward,done,b_prob,sigma])

    def clear(self):
        self.memory.clear()
        
    def sample(self, batch_size):
        tempBuf = []
        x = random.sample(range(0, len(self.memory) - self.nStep), batch_size)
        for m in x:
            tempBuf.append(list(itertools.islice(self.memory, m, m + self.nStep + 1)))
        
        return tempBuf

    def __len__(self):
        return len(self.memory)
```

```python
def train(b_net,pi_net,buffer,optimizer,batch_size,gamma,n_step,action_dim,count,soft_update):
    storedG = []
    storedS = []
    storedA = []
    experiences = buffer.sample(batch_size)
    for ex in experiences:
        #first observation will always be the current state
        states,actions,rewards,dones,bProbs,sigmas = zip(*ex)
        s_tn = states[n_step]
        a_tn = actions[n_step]
        q_tn = pi_net.forward(s_tn).detach().squeeze(0)
        G = q_tn[a_tn]
        for i in range(n_step,0,-1):
            done = dones[i]
            r_k = rewards[i]
            if done:
                G = r_k
            else:
                s_k = states[i]
                a_k = actions[i]


                bProb_k = bProbs[i]
                sigma = sigmas[i]


                pi_prob = pi_net.action(s_k)
                piProb_k = pi_prob[a_k]
                pho_k = piProb_k / bProb_k

                q_k = pi_net.forward(s_k).detach().squeeze(0)
                VBar = np.sum([(pi_prob[a]) * q_k[a] for a in range(action_dim)])
                G = r_k + gamma * ((sigma * pho_k) + ((1-sigma) * pi_prob[a_k])) * (G - q_k[a_k])+ gamma * VBar

        s_l = states[0].squeeze(0)
        a_l = actions[0]
        storedS.append(s_l)
        storedA.append(a_l)
        storedG.append(G)
    storedG = torch.tensor(storedG, dtype=torch.float32, device=device)
    storedS = torch.stack(storedS, dim=0)
    storedA = torch.tensor(storedA, dtype=torch.int64, device=device)
    
    q_values = b_net.forward(storedS)
    q_value = q_values.gather(1, storedA.unsqueeze(1)).squeeze(1)
    loss = (storedG - q_value).pow(2)
    loss = loss.mean()
    optimizer.zero_grad()
    loss.backward()

    optimizer.step()

```

```python
def chooseSigma():
     return 0.5
```

```python
episode_durations = []
def plot_durations(show_result=False):
    plt.figure(1)
    durations_t = torch.tensor(episode_durations, dtype=torch.float)
    if show_result:
        plt.title('Result')
    plt.xlabel('Episode')
    plt.ylabel('Duration')
    plt.plot(durations_t.numpy())
    # Take 100 episode averages and plot them too
    if len(durations_t) >= 100:
        means = durations_t.unfold(0, 100, 1).mean(1).view(-1)
        means = torch.cat((torch.zeros(99), means))
        plt.plot(means.numpy())
```

```python
gamma = 0.99
learning_rate = 0.0001
batch_size = 64
soft_update = 50
capacity = 10000


if torch.cuda.is_available():
    episode = 600
else:
    episode = 600

n_step = 3

buffer = replay_buffer(capacity,n_step)

env = gym.make('CartPole-v1')
observation_dim = env.observation_space.shape[0]
action_dim = env.action_space.n
actions = list(range(0,env.action_space.n))

b_epsilonS = 1
b_epsilonE = 0.1

b_net = DQN(observation_dim,action_dim,b_epsilonS,b_epsilonE).to(device)
pi_net = DQN(observation_dim,action_dim).to(device)
b_net.load_state_dict(pi_net.state_dict())

optimizer = torch.optim.Adam(b_net.parameters(), lr=learning_rate)
count = 0
for i in range(episode):
    print(i)
    T = np.inf
    t = 0
    tau = 0
    totalRew = 0
    state = env.reset()
    state = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
    b_prob = b_net.action(state)
    action = np.random.choice(actions,p=b_prob)
    b_actionProb = b_prob[action]
    done = 0
    sigma = chooseSigma()
    while t<T:
        next_state, reward, next_done, info = env.step(action)
        totalRew = totalRew + reward
        buffer.store(state,action,reward,done,b_actionProb,sigma)
        done = next_done
        state = torch.tensor(next_state, dtype=torch.float32, device=device).unsqueeze(0)
        if done:
            T=t+1
            
        else:
            b_prob = b_net.action(state)
            action = np.random.choice(actions,p=b_prob)
            b_actionProb = b_prob[action]
            sigma = chooseSigma()
        t = t+1
        
    if len(buffer) >= (batch_size + n_step):
        count = count + 1
        train(b_net,pi_net,buffer,optimizer,batch_size,gamma,n_step,action_dim,count,soft_update)
    if (count+1) % soft_update == 0:
        pi_net.load_state_dict(b_net.state_dict())
    episode_durations.append(T)


print('Complete')
plot_durations(show_result=True)

plt.savefig('results.png')
plt.show()
```
